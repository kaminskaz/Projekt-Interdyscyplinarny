```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
b0_non_cif_aug_comb = pd.read_csv('/Users/karolina/Desktop/PI/pi/EfficientNet-B0 Non-Pretrained_CIFAR10True_Augmentor_combine.csv')
b0_non_cif_aug_diff = pd.read_csv('/Users/karolina/Desktop/PI/pi/EfficientNet-B0 Non-Pretrained_CIFAR10True_Augmentor_different.csv')
b0_non_cif_aug_same = pd.read_csv('/Users/karolina/Desktop/PI/pi/EfficientNet-B0 Non-Pretrained_CIFAR10True_Augmentor_same.csv')
b0_non_cif_stes = pd.read_csv('/Users/karolina/Desktop/PI/pi/EfficientNet-B0 Non-Pretrained_CIFAR10True_STESAugmentor_None.csv')
b0_cif_base = pd.read_csv('/Users/karolina/Desktop/PI/pi/EfficientNet-B0 Non-Pretrained_CIFAR10True_NoneType_None.csv')
```

```python
def plot_metrics(models_data, titles):
    # Tworzymy subplots: 1 wiersz, 5 kolumn (po jednej dla kadej metryki)
    fig, axs = plt.subplots(1, 5, figsize=(20, 5))

    # Lista metryk
    metrics = ['accuracy', 'loss', 'f1', 'recall', 'precision']
    colors = ['blue', 'green', 'red', 'purple', 'orange']

    for i, metric in enumerate(metrics):
        # Przechodzimy przez kady wykres
        for j, (model_data, _) in enumerate(models_data):
            # Uywamy tylko krtkiej nazwy modelu w label
            axs[i].plot(model_data.index, model_data[metric], label=titles[j], color=colors[j])

        # Ustawiamy tytu i etykiety dla kadego wykresu
        axs[i].set_title(f'{metric.capitalize()}')
        axs[i].set_xlabel('Epoch')
        axs[i].set_ylabel('Value')
        axs[i].legend()
        axs[i].grid()

    plt.tight_layout()  # Aby wykresy si nie nakaday
    plt.show()

# Przykad wywoania
models_data = [
    (b0_non_cif_aug_comb, 'Model 1'),
    (b0_non_cif_aug_diff, 'Model 2'),
    (b0_non_cif_aug_same, 'Model 3'),
    (b0_non_cif_stes, 'Model 4'),
    (b0_cif_base, 'Model 5')
]

plot_metrics(models_data, titles=['Model 1', 'Model 2', 'Model 3', 'Model 4', 'Model 5'])
```
